```python
#grouping data by 'Order Date' so that all the orders that share the same date are counted in a single row - sum of the prices paid in 'Total Charged' column for each date 
```

```python
daily_orders = df.groupby('Order Date').sum()["Total Charged"]
daily_orders.head()
```

```python
#bar chart for daily orders:
```

```python
daily_orders.plot.bar(figsize=(20, 10), color='red')
plt.title("Total Charged per day")
plt.xlabel("Order Date")
plt.ylabel("Amount")
```
