```python
import matplotlib.pyplot as plt
import numpy as np

y1 = np.array([1, 2, 3, 4])
y2 = np.array([5, 6, 7, 8])
z1 = np.array([9, 10, 11, 12])

plt.plot(y1)
plt.plot(y2)
plt.plot(z1)


plt.subplot(2, 3, 1)
plt.title("Books")
plt.plot(y1,y2,z1)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 2)
plt.title("Drinks")
plt.scatter(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 3)
plt.title("Shows")
plt.bar(x, y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 4)
plt.title("Family")
plt.pie(y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 5)
plt.grid(color = 'red', linestyle = '--', linewidth = 0.9)
plt.title("Watch time")
plt.plot(x,y)

x = np.array([0, 1, 2, 3, 4, 4, 5, 1, 1, 8, 8, 8, 8, 8, 8])
y = np.array([10, 20, 30, 40])
plt.title("Time")

plt.subplot(2, 3, 6)
plt.title("Frequency")
plt.hist(x)

plt.show()
```
